#1 - Apresente em tela (output) toda a base de dados. Pontuação: 0,5 *pontos*

```python
import pandas as pd
#DADOS DAS MÚSICAS - (1º Passo)
Nomes_Musicais = ["Haverá Sinais", "leve com Você", "Gosta de Rua", "Valores", "Me namora", "Ela Só Quer Paz"]
Artistas = ["Jorge e Mateus", "Natiruts", "Felipe & Rodrigo", "Tribo Da Periferia", "Natiruts", "Projota"]
Generos = ["Sertanejo", "Reggae", "Sertanejo", "Rap", "Reggae", "Rap"]
Ano_Lancamento = [2024, 2002, 2024, 2014, 2016, 2016]
Duracao_Minutos = [3.10, 4.02, 2.51, 2.5, 4.37, 2.54]
#CRIANDO UM DICIONÁRIO CONTENDO OS DADOS DAS MÚSICAS - (2º Paaso)
dados_musicais = {      #NOME DO MEU DICIONÁRIO
    "Música": Nomes_Musicais,
    "Artista": Artistas,
    "Gêneros": Generos,
    "Lançamento": Ano_Lancamento,
    "Duração": Duracao_Minutos
}

#CRIANDO UM DATAFRAME A PARTIR DO DICIONARIO - (3º Passo)
minha_playlist = pd.DataFrame(dados_musicais)
display(minha_playlist)
```

#2 - Apresente o tamanho do seu dataframe (quantas colunas x linhas). Pontuação: 0,5 pontos

```python
tamanho = minha_playlist.shape
display("Número de linhas:", tamanho[0])  # Acessando primeiro elemento da tupla, que representa o número de linhas. Indexação em Python começa em 0
display("Número de colunas:", tamanho[1])
```

#6 - Exclua um item (linha) de sua base de dados. Pontuação: 0,5 pontos

```python
# Exibindo o DataFrame original
print("DataFrame original:")
display(minha_playlist)
# Excluindo a terceira linha (índice 2)
minha_playlist = minha_playlist.drop(2).reset_index(drop=True)
# Exibindo o DataFrame atualizado
print("\nDataFrame após excluir a terceira linha:")
display(minha_playlist)
```

#7 - Adicione um item (linha) na sua base de dados. Pontuação: 0,5 pontos

```python
# Dados da nova música
nova_musica = {
    "Música": "Água de Beber",
    "Artista": "Tom Jobim",
    "Gênero": "Bossa Nova",
    "Lançamento": 1963,
    "Duração": 2.29
}

# Adicionando a nova música ao DataFrame usando loc
minha_playlist.loc[len(minha_playlist)] = nova_musica

# Exibindo o DataFrame atualizado
print("\nDataFrame após adicionar a nova música:")
display(minha_playlist)
```

#8 - Transponha a coluna para a linha em sua base de dados. Pontuação: 0,5 pontos
